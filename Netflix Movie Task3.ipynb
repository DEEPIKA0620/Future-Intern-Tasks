```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
%matplotlib inline
```

```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df
```

```python
print("dataset shape:",df.shape)
print("columns:",df.columns)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna(subset=['type','country','release_year','rating'])
df_clean
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df_clean, x='type')
plt.title('Number of Movies vs TV show')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
top_countries=df_clean['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, color='Green')
plt.title('Top 10 countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('country')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df_clean, y='rating',order=df_clean['rating'].value_counts().index, color='red')
plt.title('Distribution of content Ratings')
plt.xlabel('count')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df_clean, x='release_year', bins=30, kde=False, color='purple')
plt.title('Distributionf Content Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```
